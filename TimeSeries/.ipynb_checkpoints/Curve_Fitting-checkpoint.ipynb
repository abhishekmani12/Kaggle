**Decomposing the series:**

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomp=seasonal_decompose(df, period=12)  #( dataset, seasonality period=12 as df has monthly data)
figure=decomp.plot()
plt.show()

```

Residual across the years is different - negative, 0 then positive

**Using a multiplicative Model to test our theory:**

```python
decomp=seasonal_decompose(df['value'], model='multiplicative', period=12)
figure=decomp.plot()
plt.show()
```

Residual now resembles white noise (symmetrical) which indicates this series is a multiplicative model.

**Autocorrelation Analysis:**

**ACF:** Direct + Indirect

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df['value'])
print()
```

x axis: Correlation [ range: -1 to 1 ]

**Exponential Smoothing:**

TS forecasting method only for **Univariate Data**

Not a probabilistic model; Only fits curves

```python
cutoff='2005-12-31'

df['value']/=100 #normalize to solve error

#train and val split

xtrain, xval=df.loc[df.index <= cutoff], df.loc[df.index > cutoff]
print(xtrain.shape, xval.shape)
```

```python
 from statsmodels.tsa.holtwinters import ExponentialSmoothing

fit1= ExponentialSmoothing(xtrain['value'].values, seasonal_periods=12, trend='mul', seasonal='mul')

#(data, period, trend-multiplicative, seasonal-multiplicative)

fit1=fit1.fit()
```

Forecasting:

```python
pred=fit1.forecast(len(xval)).copy()

xresidual= xval['value']-pred
```

Params:

```python
fit1.params_formatted

#alpha - const for smoothing level 
#beta - smoothing trend
#gamma - smoothing seasonality

#12 seasonal components
```

**ACF plot for X residuals**

```python
plot_acf(xresidual)
print()
```

concetrate on lags before 7

**PACF Plot:**

```python
plot_pacf(xresidual)
print()
```

Indicates more correlation (t) between current observation and the immediate previous observation (t-1) rather than (t-2) and so on

```python
df=pd.read_csv('/content/GreenhouseClimate1.csv')
df
```

Target: 'Tair'

Formatting date:

```python
df['time']=pd.to_datetime(df['time'], unit='d', origin='1899-12-30') #epoch conversion
df.head()
```

Time increments of every 5 minutes

Removing unwanted columns:

```python
cdf=df.copy()
cols=['time', 'Tair', 'co2_sp', 'dx_sp', 't_rail_min_sp', 't_grow_min_sp',
        'assim_sp', 'scr_enrg_sp', 'scr_blck_sp', 't_heat_sp',
        't_vent_sp', 'window_pos_lee_sp', 'water_sup_intervals_sp_min',
        'int_blue_sp', 'int_red_sp', 'int_farred_sp',
        'int_white_sp']

df=df[cols]        
```

**Resampling data:** From minutes to hours:

Frequency conv

```python
df0=df.resample('H', on='time').mean().reset_index()[cols]
```

Remove Nan:

```python
import numpy as np

np.isnan(df0[cols]).describe()
```

```python
df0.dropna(subset=['Tair'], inplace=True)
```

**Trend Plot:**

```python
df0['Tair'][0:700].plot()
```

**6 months of hourly data**

**HardCoding TimeStamp and Target Column:** As Prohphet requires it

Timestamp: 'ds'

Target: 'y'

```python
df1=df0[['time', 'Tair']].rename(columns={'time':'ds', 'Tair':'y'})
```

```python
pm=Prophet(
    weekly_seasonality=False, #Dataset has weeks but this seasonality is not required for the target var as it changes over a longer time period rather than a short one
    interval_width=0.95, #95% uncertainty/confidence coverage of preds
    mcmc_samples=10 #param estimation

)
```

**Train:**

```python
pm.fit(df1)
```

**Val:**

```python
future=pm.make_future_dataframe(periods=24, freq='H')
preds=pm.predict(future)
pm.plot_components(preds)
print()
```

**Without mcmc:**

```python
pm=Prophet(
    weekly_seasonality=False, #Dataset has weeks but this seasonality is not required for the target var as it changes over a longer time period rather than a short one
    interval_width=0.95, #95% uncertainty/confidence coverage of preds
   

)
pm.fit(df1)
future=pm.make_future_dataframe(periods=24, freq='H')
preds=pm.predict(future)
pm.plot_components(preds)
print()
```

**Metrics for Val results**:

```python
from fbprophet.diagnostics import cross_validation, performance_metrics

metrics_df = cross_validation(pm, initial='3700 hours', period='24 hours', horizon='24 hours')

metrics_df
```

ds - time

y - actual value

yhat - prediction

yhat_lower / yhat_upper - confidence range

cutoff - End of time period used for the specified batch inference


**Aggregated Metrics:**

For better clarity

```python
agg_df=performance_metrics(metrics_df)
agg_df
```

**Feature Selection:**

```python
from sklearn.linear_model import Ridge
from sklearn.feature_selection import RFE
```

```python
df0=df0.dropna()

cols  = ['co2_sp', 'dx_sp', 't_rail_min_sp', 't_grow_min_sp',
        'assim_sp', 'scr_enrg_sp', 'scr_blck_sp', 't_heat_sp',
        't_vent_sp', 'window_pos_lee_sp', 'water_sup_intervals_sp_min',
        'int_blue_sp', 'int_red_sp', 'int_farred_sp',
        'int_white_sp']

estimator= Ridge()
selector=RFE(estimator, n_features_to_select=3, step=1)

selector=selector.fit(df0[cols], df0['Tair'])
 
imp_cols= df0[cols].columns[selector.support_]

print(imp_cols)
```

```python
#Renaming

df3=df0[['time','Tair']].rename(columns={'time' : 'ds', 'Tair': 'y'})

#Insantiation
mp=Prophet(weekly_seasonality=False, interval_width=0.95)
```

```python
#Adding Regressors(imp_cols) to df3
for col in imp_cols:

  df3[col]=df0[col]
  mp.add_regressor(col)

#Train

mp.fit(df3)
```

**Val and Metrics:**

```python
metrics_df = cross_validation(mp, initial='3700 hours', period='24 hours', horizon='24 hours')
agg_df1=performance_metrics(metrics_df)
```

```python
agg_df1
```

**Comparison b/w univariate and covariates:**

```python
comparison=pd.DataFrame()
comparison['uni    ']=agg_df.mean(axis=0)[1:]
comparison['  Co']=agg_df1.mean(axis=0)[1:]

print(comparison)
```
